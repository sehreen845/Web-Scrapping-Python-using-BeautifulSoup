```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com/")
web
```

```python
soup = BeautifulSoup(web.content,"html.parser")
print(soup.prettify())
```

```python
soup.find_all('p')
```

```python
soup.find_all('a')
```

```python
soup.find_all('h1')
```

```python
class_data = soup.find("div",class_="container")
class_data
```
